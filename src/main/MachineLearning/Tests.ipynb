# Input Size

We want to identify whether a different size of the input image will have a significant effect on the accuracy of the bird classication task. We have decided to test this by regarding the output of isBird before and after resizing an image.

```python
confusion_matrix = [[0,0],[0,0]]
for directory in listdir('./data/'):
    if directory not in ["Grass","Leaf","House"]:
        print(f"We enter directory {directory}")
        for (index,file) in enumerate(listdir('data/' + directory),0):
            if index > 9 and index < 15:
                real = isBird('./data/' + directory + '/' + file)[0]
                print(real)
                IMG = Image.open('data/' + directory + '/' + file)
                IMG = IMG.resize((512,512))
                IMG.save('data/' + directory + '/' + file)
                pred = isBird('data/' + directory + '/' + file)[0]
                print(pred)
                confusion_matrix[real][pred] = confusion_matrix[real][pred] + 1
print(confusion_matrix)
```

In the 2D confusion matrix, we see that all results are True Positives, so all images were correctly classified both before and after resizing.

This shows that the model can handle various sizes of the input image. So we do not need to resize images. 

On the other hand, we could resize the images to speed up the isBird function because posting larger images takes more time.

# Blurry condensation

We want to identify whether blurring the input image will have a significant effect on the accuracy of the bird classication task. We have decided to test this by regarding the output of isBird before and after blurring an image.

```python
confusion_matrix = [[0,0],[0,0]]
for directory in listdir('./data/'):
    if directory not in ["Grass","Leaf","House"]:
        print(f"We enter directory {directory}")
        for (index,file) in enumerate(listdir('data/' + directory),0):
            if index < 5:
                real = isBird('./data/' + directory + '/' + file)[0]
                print(real)
                IMG = Image.open('data/' + directory + '/' + file)
                IMG = IMG.filter(ImageFilter.GaussianBlur(radius=5))
                IMG.save('data/' + directory + '/' + file)
                pred = isBird('data/' + directory + '/' + file)[0]
                print(pred)
                confusion_matrix[real][pred] = confusion_matrix[real][pred] + 1
print(confusion_matrix)
```

In the 2D confusion matrix, there is a large value in the right bottom entry and low values in the other entries. Out of 100 images, 98 images were classified correctly before and after blurring. We misclassified one image of a Collared Dove twice (before and after blurring). We missclassified one image of a WoodPigeon (after blurring).

This shows that the model is robust to blurring. So the model will do well at detecting birds in images even when there is condensation.

# Artificial rain

We want to identify whether adding some noise in the form of rain-like patterns to the input image will have a significant effect on the accuracy of the bird classication task. We have decided to test this by regarding the output of isBird before and after adding this noise to an image.

```python
def generate_random_lines(imshape,slant,drop_length):
  drops=[]
  for _ in range(1500): ## If You want heavy rain, try increasing this
    if slant<0:
      x = randint(slant,imshape[1])
    else:
      x = randint(0,imshape[1]-slant)
    y = randint(0,imshape[0]-drop_length)
    drops.append((x,y))
  return drops

def add_rain(image):
  imshape = image.size
  slant_extreme = 10
  slant= randint(-slant_extreme,slant_extreme)
  drop_length=20
  drop_width=2
  drop_color=(200,200,200) ## a shade of gray
  rain_drops= generate_random_lines(imshape,slant,drop_length)
  for rain_drop in rain_drops:
    draw = ImageDraw.Draw(image)
    draw.line((rain_drop[0],rain_drop[1], rain_drop[0]+slant,rain_drop[1]+drop_length), fill=drop_color, width=drop_width)
  image = image.filter(ImageFilter.GaussianBlur(radius=1))
  filter = ImageEnhance.Brightness(image)
  image = filter.enhance(0.70)
  return image
```

```python
confusion_matrix = [[0,0],[0,0]]
for directory in listdir('./data/'):
    if directory not in ["Grass","Leaf","House"]:
        print(f"We enter directory {directory}")
        for (index,file) in enumerate(listdir('data/' + directory),0):
            if index > 4 and index < 10:
                real = isBird('./data/' + directory + '/' + file)[0]
                print(real)
                IMG = Image.open('data/' + directory + '/' + file)
                IMG = add_rain(IMG)
                IMG.save('data/' + directory + '/' + file)
                pred = isBird('data/' + directory + '/' + file)[0]
                print(pred)
                confusion_matrix[real][pred] = confusion_matrix[real][pred] + 1
print(confusion_matrix)
```

In the 2D confusion matrix, there is a large value in the right bottom entry and low values in the other entries. Out of 100 images, 70 images were classified correctly before and after blurring. We misclassified 30 images and most of those misclassifications happened because the addition of rain effects made it more difficult to detect a bird in the image. 

This shows that the model is quite robust to slight blurring, brightness lowering, and the addition of rain-like grey lines. So the model will do well at detecting birds in images even when the outdoor environment is rainy.
